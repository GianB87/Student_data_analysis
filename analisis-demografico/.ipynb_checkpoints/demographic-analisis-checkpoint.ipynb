## Outlier
* It is evident that there is an outlier due to a bad enter of data

```python
df[df.columns[0]].value_counts().plot.pie()
```

```python
df[df.columns[0]].plot.box()
```

```python
# Identifying outlier
location = np.where(df['indique su edad']>5000)[0]
df['indique su edad'][location] 

# Replacing to 22 (probably the desired value)
df["indique su edad"] = df["indique su edad"].mask(df["indique su edad"] >75, 22)
```
